```python
def distribution_function(column_name):
    """
    This function calculates the distribution of values in the specified column,
    generates a horizontal bar plot representing the distribution, and displays it
    
    """
    
    
    var_crosstab = pd.crosstab(index=df.loc[:, column_name], columns='count', normalize='columns')*100
    var_crosstab_sorted = var_crosstab.iloc[:, 0].sort_values()
    display(var_crosstab)
    fig, ax = plt.subplots(figsize=(6, 6))
    var_crosstab_sorted.plot(kind='barh', ax=ax)
    ax.set_xticklabels(['{:.0f}%'.format(x) for x in ax.get_xticks()]);
```

```python
def attrition_function(column_name):
    """
    This function filters the DataFrame by 'Attrition' where the value is 'Yes', calculates
    the distribution of values in the specified column for the filtered data, generates a
    horizontal bar plot representing the distribution, and displays it.

    
    """
     
    filtered_df = df[df['Attrition'] == 'Yes']
    var_crosstab = pd.crosstab(index=filtered_df.loc[:, column_name], columns='count', normalize='columns') * 100
    var_crosstab_sorted = var_crosstab.iloc[:, 0].sort_values()
    display(var_crosstab)
    fig, ax = plt.subplots(figsize=(6, 6))
    var_crosstab_sorted.plot(kind='barh', ax=ax)

    ax.set_xticklabels(['{:.0f}%'.format(x) for x in ax.get_xticks()])
```

```python
# define function to show distributions of categorial variable (+ filtered by attrition)
def distribution_function(column_name):
    '''
    This function takes one argument, namely the column name from a 'df'-called DataFrame,
    and returns the value distribution of the selected column, the attrition rate for the values of the selected column,
    and a figure with one plot for the distribution and one for the attrition rate
    '''
    var_crosstab = pd.crosstab(index=df.loc[:, column_name], columns='count', normalize='columns')*100
    var_crosstab_sorted = var_crosstab.iloc[:, 0].sort_values()
    display(var_crosstab)
    
    var_crosstab_att = pd.crosstab(index=df.loc[:, column_name], columns=df.loc[:, 'Attrition'], normalize='index')*100
    var_crosstab_att_sorted = var_crosstab_att.sort_values(by='Yes', ascending=False)
    display(var_crosstab_att_sorted)
   
    fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(12, 12))
    var_crosstab_sorted.plot(kind='barh', ax=ax[0])
    ax[0].set_xticklabels(['{:.0f}%'.format(x) for x in ax[0].get_xticks()])
    var_crosstab_att_sorted.plot(kind='barh', legend=True, ax=ax[1])
    ax[1].set_xticklabels(['{:.0f}%'.format(x) for x in ax[1].get_xticks()]);
```

```python
# define function to show most important statistics and distribution of numerical variables
def stats_and_distribution(column_name):
    '''
    This function takes one argument, namely the column name from a 'df'-called DataFrame,
    and returns first the most relevant descriptive statistics for the chosen column plus the same outpot grouped by Attrition,
    and second two figures with two axes.
       
       first figure:
        - a plotted histogram of the selected DataFrame-column on the first axis
        - a plotted boxplot of the selected DataFrame-column on the second axis
       
       second figure:
        - a plotted histogram of the selected DataFrame-column on the first axis grouped by attrition
        - a plotted boxplot of the selected DataFrame-column on the second axis grouped by attrition
    '''
    display(df.loc[:, column_name].describe())
    display(df.groupby('Attrition')[column_name].describe().T)
    
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (12, 6))
    df.loc[:, column_name].plot(kind='hist', legend=True, ax=ax[0]);
    df.boxplot(column=column_name, showfliers=True, ax=ax[1]);
    
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (12, 6))
    df.groupby('Attrition')[column_name].plot(kind='hist', legend=True, ax=ax[0])
    df.boxplot(column=column_name, by='Attrition', showfliers=True, ax=ax[1]);
```

# 3.3.1 Attrition Rate of nominal variables

```python
def calculate_and_sort_percentages(df):
    """
    This function calculates the percentages of each category within categorical columns with respect to the target variable.
    It then sorts these percentages in descending order based on the target category ('Attrition' = 1) and stores the results
    in a dictionary with column names as keys.
    
    """
    categorical_columns = ['AgeGroup', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'SalarySlab', 'OverTime']
    target_column = 'Attrition'

    results = {}
    for column in categorical_columns:
        # crosstabs
        percentages = pd.crosstab(index=df[column], columns=df[target_column], normalize='index') * 100
        # sort
        sorted_percentages = percentages.sort_values(by="1", ascending=False)
        results[column] = sorted_percentages
    return results

# call the function
results = calculate_and_sort_percentages(df)

for column, result in results.items():
    display(result)
```

```python
# Daten einfgen
var_cross_tab = pd.crosstab(index=df.loc[:, 'AgeGroup'],
                            columns=df.loc[:, 'Attrition'],
                            normalize='index')

# Durchschnittliche Attrition-Rate fr den gesamten Datensatz berechnen
total_average_attrition_rate = df['Attrition'].astype(int).mean()

# Balken absteigend nach Attrition-Rate sortieren
var_cross_tab_sorted = var_cross_tab.sort_values(by='1', ascending=False)

# Plot erstellen
ax = var_cross_tab_sorted.plot(kind='bar', figsize=(8,5), ylim=(0,1.15),
                               title='Age Group and Attrition')

# Durchschnittliche Attrition-Rate fr den gesamten Datensatz als gestrichelte Linie einfgen
ax.axhline(y=total_average_attrition_rate, color='orange')

# Y-Achse in Prozent anzeigen
ax.set_yticklabels(['{:,.0%}'.format(x) for x in ax.get_yticks()])

# Plot anzeigen
plt.show()
```

# 3.3.3 Attrition Probability of metric variables

```python
# check intercorrelations
scaled_metrically = ['Age', 'MonthlyIncome', 'TotalWorkingYears', 
                     'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
sns.heatmap(df[scaled_metrically].corr(), annot=True);
```

```python
for entry in scaled_metrically:
    print("\033[1m" + entry + "\033[0m")
    display(df.groupby('Attrition')[entry].describe())
    print('')
```

```python
df['Attrition'] = df['Attrition'].astype('int16')

import math

# define function for logistic regression, calculation threshold and plotting the model and threshold
def logistic_regression_and_plot(metric_variable, suitable_name):
    '''
    This function takes two arguments, namely the column name (if metrically scaled) from a 'df'-called DataFrame
    and a suitable name for what the column stands for (e.g. MonthlyIncome = Monthly Income).
    
    The function returns a logistic regression of the attrition probability in dependence of the selected variable,
    then plots the model including a threshold line at that threshold 
    where the attrition probability is lower than the average attrition rate.
    '''
    
    model = smf.logit(formula='Attrition ~{}'.format(metric_variable), data=df)
    results = model.fit()
    display(results.summary())

    X = pd.Series(range(df.loc[:, metric_variable].max()+1))
    intercept = results.params['Intercept']
    slope = results.params[metric_variable]
    p_y = 1/(1+math.e**(-(intercept+(slope*X))))*100
    print(p_y, '\n')

    # attrition probability lower than the average attrition rate of 16.12% 
    # (change operator if direction of correlation is the other way around) 
    threshold = p_y[p_y <= 16.12].index[0]
    print(threshold, '\n')
    
    # plotting regression model
    plt.style.use('fivethirtyeight')
    fig, ax = plt.subplots(figsize=(12, 8))
    p_y.plot(title='Attrition Probability by {}'.format(metric_variable), ax=ax)
    ax.vlines(x=threshold, ymin=0, ymax=100, colors='orange')
    ax.set(xlabel=metric_variable, ylabel='Attrition Probability')
    ax.set_yticklabels(['{:.0f}%'.format(x) for x in ax.get_yticks()])
    
    ann_arrow = ax.annotate(text='''From {} = {} on: attrition probability is 
lower than the average attrition rate'''.format(suitable_name, threshold),
                       xy=[0, 0],
                       xytext=[threshold*1.05, 70]);
```

```python
# Final Visualization (Age)
model = smf.logit(formula='Attrition ~ Age', data=df)
results = model.fit()
display(results.summary())

X = pd.Series(range(df.loc[:, 'Age'].max()+1))
intercept = results.params['Intercept']
slope = results.params['Age']
p_y = 1/(1+math.e**(-(intercept+(slope*X))))*100
print(p_y, '\n')

# attrition probability lower than the average attrition rate of 16.12% 
# (change operator if direction of correlation is the other way around) 
threshold = p_y[p_y <= 16.12].index[0]
print(threshold, '\n')
    
# plotting regression model
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12, 8))
p_y.plot(title='Je jnger die Mitarbeiter, desto eher verlassen sie das Unternehmen wieder', ax=ax)
ax.vlines(x=threshold, ymin=0, ymax=100, colors='orange')
ax.set(xlabel='Alter', ylabel='Abwanderungswahrscheinlichkeit')
ax.set_yticklabels(['{:.0f}%'.format(x) for x in ax.get_yticks()]);
    
ann_arrow = ax.annotate(text='''Ab einem Alter von 36 Jahren
liegt die Abwanderungswahrscheinlichkeit
der Mitarbeiter unter der 
durchschnittlichen Abwanderungsrate.''',
                      xy=[0, 0],
                       xytext=[threshold*1.05, 18]);
```

```python
# Final Visualization (MonthlyIncome)
model = smf.logit(formula='Attrition ~ MonthlyIncome', data=df)
results = model.fit()
display(results.summary())

X = pd.Series(range(df.loc[:, 'MonthlyIncome'].max()+1))
intercept = results.params['Intercept']
slope = results.params['MonthlyIncome']
p_y = 1/(1+math.e**(-(intercept+(slope*X))))*100
print(p_y, '\n')

# attrition probability lower than the average attrition rate of 16.12% 
# (change operator if direction of correlation is the other way around) 
threshold = p_y[p_y <= 16.12].index[0]
print(threshold, '\n')
    
# plotting regression model
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12, 8))
p_y.plot(title='Je weniger die Mitarbeiter verdienen, desto eher verlassen sie das Unternehmen wieder', ax=ax)
ax.vlines(x=threshold, ymin=0, ymax=100, colors='orange')
ax.set(xlabel='Gehalt', ylabel='Abwanderungswahrscheinlichkeit')
ax.set_yticklabels(['{:.0f}%'.format(x) for x in ax.get_yticks()])
ax.set_xticklabels(['{:.0f}$'.format(x) for x in ax.get_xticks()]);
    
ann_arrow = ax.annotate(text='''Ab einem Gehalt von 5.567$
liegt die Abwanderungswahrscheinlichkeit
der Mitarbeiter unter der 
durchschnittlichen Abwanderungsrate.''',
                      xy=[0, 0],
                       xytext=[threshold*1.05, 18]);
```

```python
# logistic regression: Attrition ~ TotalWorkingYears
logistic_regression_and_plot('TotalWorkingYears', 'years of work experience');
```

```python
# logistic regression: Attrition ~ YearsAtCompany
logistic_regression_and_plot('YearsAtCompany', 'years at the company');
```

```python
# logistic regression: Attrition ~ YearsInCurrentRole
logistic_regression_and_plot('YearsInCurrentRole', 'years in current role');
```

```python
# logistic regression: Attrition ~ YearsWithCurrManager
df['YearsWithCurrManager'] = df['YearsWithCurrManager'].astype('int16') 
logistic_regression_and_plot('YearsWithCurrManager', 'years with current manager');
```
